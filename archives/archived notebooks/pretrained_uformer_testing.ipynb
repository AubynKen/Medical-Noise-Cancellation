```python
import torch
import numpy as np
from PIL import Image
from torchvision import transforms
import copy
```

```python
inp_path = "tests/input/stent.png"
gt_path = "tests/groundtruth/stentgt.png"
out_path = "tests/output/stentgt.png"

# Assure test image resolution is 1024x1024
inp_img = Image.open(inp_path)
if inp_img.size != (1024, 1024):
    inp_img = inp_img.resize((1024,1024))
    inp_img.save(inp_path)
gt_img = Image.open(gt_path)
if gt_img.size != (1024, 1024):
    gt_img = gt_img.resize((1024, 1024))
    gt_img.save(gt_path)
```
